```python
import matplotlib.pyplot as plt
```

```python
def payoff(price: float):
    deposited_value = 300 * price
    borrowed = 600_000
    if borrowed/deposited_value > 0.83:
        return 5_000
    if deposited_value - borrowed > 200_000:
        return 200_000
    return deposited_value - borrowed
```

```python
_curve = [payoff(price) for price in range(1, 6000)]
```

```python
plt.plot(_curve)
```

```python
payoff(4000)
```
